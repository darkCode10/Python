**Greedy Best-First Search**

```python
import heapq

# Sample graph as a dictionary with path costs
graph = {
    'A': {'B': 1, 'C': 2},
    'B': {'D': 4, 'E': 5},
    'C': {'F': 3, 'G': 6},
    'D': {},
    'E': {},
    'F': {},
    'G': {}
}

# Heuristic values for each node
heuristic = {
    'A': 6,
    'B': 4,
    'C': 2,
    'D': 7,
    'E': 3,
    'F': 2,
    'G': 0  # The heuristic for the goal is always 0
}

def greedy_best_first_search(graph, start, goal):
    priority_queue = [(heuristic[start], start, [])]  # Priority queue of (heuristic, node, path)
    visited = set()

    while priority_queue:
        h, current, path = heapq.heappop(priority_queue)  # Get node with the lowest heuristic
        path = path + [current]

        if current == goal:  # Check if reached goal
            return path

        if current not in visited:
            visited.add(current)
            for neighbor, path_cost in graph.get(current, {}).items():
                if neighbor not in visited:
                    heapq.heappush(priority_queue, (heuristic[neighbor], neighbor, path))  # Add neighbor with its heuristic
    return None

start_node = 'A'
goal_node = 'D'

gbfs_path = greedy_best_first_search(graph, start_node, goal_node)
if gbfs_path:
    print(f"GBFS Path: {gbfs_path}")
else:
    print("No path found.")
```

**A* Search**

```python
import heapq

# Sample graph as a dictionary with path costs
graph = {
    'A': {'B': 1, 'C': 2},
    'B': {'D': 4, 'E': 5},
    'C': {'F': 3, 'G': 6},
    'D': {},
    'E': {},
    'F': {},
    'G': {}
}

# Heuristic values for each node
heuristic = {
    'A': 6,
    'B': 4,
    'C': 2,
    'D': 7,
    'E': 3,
    'F': 2,
    'G': 0  # The heuristic for the goal is always 0
}

def a_star_search(graph, start, goal):
    cost = 0
    path = []
    priority_queue = [(heuristic[start], cost, start, path)]  # Priority queue of (estimated_cost, actual_cost, node, path)
    visited = set()

    while priority_queue:
        est_cost, actual_cost, current, path = heapq.heappop(priority_queue)
        path = path + [current]

        if current == goal:  # Check if reached goal
            return actual_cost, path

        if current not in visited:
            visited.add(current)
            for neighbor, path_cost in graph.get(current, {}).items():
                if neighbor not in visited:
                    new_actual_cost = actual_cost + path_cost # Calculate new actual cost to this neighbor (g)
                    est_cost = new_actual_cost + heuristic[neighbor] # Estimated total cost (f = g + h)
                    heapq.heappush(priority_queue, (est_cost, new_actual_cost, neighbor, path)) # Add neighbor to priority queue

    return None, float('inf')

start_node = 'A'
goal_node = 'G'

a_star_cost, a_star_path = a_star_search(graph, start_node, goal_node)
if a_star_path:
    print(f"A* Path: {a_star_path} with cost: {a_star_cost}")
else:
    print("No path found.")
```
